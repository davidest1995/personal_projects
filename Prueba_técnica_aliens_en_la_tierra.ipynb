```python
initial_state = [3,4,3,1,2]

def days_counter(days):
  for day in range(0,days):
    discount_day(initial_state)

def discount_day(initial_state):
  for idx,element in enumerate(initial_state):
    if initial_state[idx] == 0:
      initial_state[idx] = 7
      initial_state.append(9) 
    initial_state[idx] = initial_state[idx] - 1
  print(initial_state)


if __name__ == '__main__':
  print('el initial state es : {}'.format(initial_state))
  days= int(input('cuantos días después quiere ver el estado de los aliens? '))
  print('La cantidad de aliens, pasados {} días es : '.format(days))
  days_counter(days)
```
